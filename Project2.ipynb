Formulation:
$\begin{equation}
-v(t) + R_1 i(t) + R_2 i(t) = 0 \Rightarrow i(t) = \frac{1}{R_1 + R_2} v(t)
\end{equation}$

$\begin{equation}
v_{out}(t) = \frac{1}{1 + R_1 / R_2} v(t) \Rightarrow \mid v_{out}(t) \mid < \mid v(t) \mid
\end{equation}$

The Eq. (2) tells the output signal is less than the input one, so we can combine the resistances $R_1$ and $R_2$ in a way to reduce the original signal into the desired one. For the special case $R_1 = R_2 \equiv R$ we have $\mid v_{out}(t) \mid = \frac{\mid v(t) \mid}{2}$

The current is in phase with the voltage input and output.

## Methods for solving liner systems
### Inverting the coefficient matrix
$\begin{equation}
Ax = b \Rightarrow x = A^{-1}b
\end{equation}$

```python
from numpy import array, eye

# Inverting matrix:


def gauss_inv(A, b):
    N = A.shape[0]
    A_inv = eye(N)
    for i in range(N):
        div = A[i, i]
        for k in range(N):
            A[i, k] /= div
            A_inv[i, k] /= div
        for l in range(N):
            if l != i:
                mult = A[l, i]
                for m in range(N):
                    A[l, m] -= mult*A[i, m]
                    A_inv[l, m] -= mult*A_inv[i, m]
    x = A_inv.dot(b)
    return x
```

### Gaussian Elimination (without pivoting)

```python
from numpy import array, empty

# Backsubstitution:


def backsubs(A, b):
    N = b.size
    x = empty(shape=(N, 1), dtype=float)
    for m in range(N-1, -1, -1):
        x[m] = b[m]
        for i in range(m+1, N):
            x[m] -= A[m, i]*x[i]
    return x

# Gaussian elimination:


def gauss_elim(A, b):
    N = b.size
    for m in range(N):
        div = A[m, m]
        for l in range(N):
            A[m, l] /= div
        b[m] /= div
        for i in range(m+1, N):
            mult = A[i, m]
            for k in range(N):
                A[i, k] -= mult*A[m, k]
            b[i] -= mult*b[m]
    x = backsubs(A, b)
    return x
```

### Tri-diagonal of text-book of Mark Newman

```python
def tri_diagonal(A, b):
    N = b.size
    for i in range(N-1):
        A[i, i+1] /= A[i, i]
        b[i] /= A[i, i]
        A[i+1, i+1] -= A[i+1, i]*A[i, i+1]
        b[i+1] -= A[i+1, i]*b[i]
    b[N-1] /= A[N-1, N-1]

    x = empty(shape=(N, 1), dtype=float)
    x[N-1] = b[N-1]
    for i in range(N-2, -1, -1):
        x[i] = b[i] - A[i, i+1]*x[i+1]
    return x
```

### Tridiagonal Matrix Algorithm (TDMA) or Thomas Algorithm

```python
from numpy import array, zeros


def TDMA(A, b):
    n = b.size
    c_temp = zeros(shape=(n, 1), dtype=float)
    d_temp = zeros(shape=(n, 1), dtype=float)
    x = zeros(shape=(n, 1), dtype=float)

    c_temp[0, 0] = A[0, 1] / A[0, 0]
    d_temp[0, 0] = b[0, 0] / A[0, 0]

    for i in range(1, n-1):
        c_temp[i, 0] = A[i, i+1] / (A[i, i] - A[i, i-1] * c_temp[i-1, 0])
        d_temp[i, 0] = (b[i, 0] - A[i, i-1] * d_temp[i-1, 0]) / \
            (A[i, i] - A[i, i-1] * c_temp[i-1, 0])

    d_temp[n-1, 0] = (b[n-1, 0] - A[n-1, n-2] * d_temp[n-2, 0]) / \
        (A[n-1, n-1] - A[n-1, n-2] * c_temp[n-2, 0])

    x[n-1, 0] = d_temp[n-1, 0]

    for i in range(n-2, -1, -1):
        x[i, 0] = d_temp[i, 0] - c_temp[i, 0] * x[i+1, 0]

    return x
```

### LU decomposition (without pivoting)

```python
from numpy import array, eye, empty

# Forward substitution:


def forw_subs(L, b):
    N = b.size
    y = empty(shape=(N, 1), dtype=float)
    # Solving Ly = b:
    for k in range(N):  # k = 0, 1, ..., N-1
        s = 0
        for l in range(k):  # l = 0, 1, ..., N-2
            s += L[k, l]*y[l]
        y[k] = (b[k] - s) / L[k, k]
    return y

# Backwards substitution:


def back_subs(U, y):
    N = y.size
    x = empty(shape=(N, 1), dtype=float)
    # Solving Ux = y:
    for k in range(N-1, -1, -1):  # k = N-1, N-2, ..., 0
        s = 0
        for l in range(k+1, N):  # l = k+1, k+2, ..., N-2, N-1
            s += U[k, l]*x[l]
        x[k] = (y[k] - s) / U[k, k]
    return x

# LU decomposition:


def LU_decom(A, b):
    N = b.size
    U = array(A, copy=True)  # U = A
    L = eye(N, dtype=float)  # Identity I_N
    # Building L, U:
    for j in range(N-1):  # j = 0, 1, ..., N-2
        if U[j, j] == 0:
            print('There is no LU decomposition for this matrix')
            break
        else:
            for i in range(j+1, N):
                mult = U[i, j]/U[j, j]
                for k in range(N):  # k = 0, 1, ..., N-1
                    U[i, k] -= mult*U[j, k]
                    L[i, k] += mult*L[j, k]
    # Solving Ly=b:
    y = forw_subs(L, b)
    # Solving Ux = y
    x = back_subs(U, y)
    return x
```

### Gauss-Jacobi

```python
from numpy import array
from numpy.linalg import norm


def gauss_jacobi(A, b, x, target):
    N = b.size
    N_max = 100
    residue = 1.0
    count = 0
    while (residue > target) and (count < N_max):
        for i in range(N):
            s = 0
            for j in range(N):
                if j != i:
                    s += A[i, j]*x[j, 0]
            x[i, 0] = (b[i, 0] - s) / A[i, i]
        residue = norm(b - A.dot(x))
        count += 1
    return x
```

### Gauss-Seidel

```python
from numpy import array
from numpy.linalg import norm


def gauss_seidel(A, b, x, target):
    N = b.size
    N_max = 100
    residue = 1.0
    count = 0
    while (residue > target) and (count < N_max):
        for i in range(N):
            s1 = 0
            x_new = array(x, copy=True)
            for j1 in range(i):  # New approximations
                s1 += A[i, j1]*x_new[j1, 0]
            s2 = 0
            for j2 in range(i+1, N):  # Old approximations
                s2 += A[i, j2]*x[j2, 0]
            x[i, 0] = (b[i, 0] - s1 - s2) / A[i, i]
        residue = norm(b - A.dot(x))
        count += 1
    return x
```

### Steepest descent

```python
from numpy import array
from numpy.linalg import norm


def steepest_descent(A, b, x, target):
    r = b - A.dot(x)
    p = array(r, copy=True)
    k = 0
    N_max = 100
    while (norm(r) > target) and (k < N_max):
        q = array(A.dot(p), copy=True)
        alpha = float((p.T).dot(r)/((p.T).dot(q)))
        x += alpha*p
        r -= alpha*q
        p = array(r, copy=True)
        k += 1
    return x
```

### Conjugate-Gradient

```python
from numpy import array
from numpy.linalg import norm


def conjugate_gradient(A, b, x, target):
    N_max = 100
    r = b - A.dot(x)
    p = array(r, copy=True)
    v = (r.T).dot(r)[0, 0]
    k = 0
    while (norm(r) > target) and (k < N_max):
        q = A.dot(p)
        m = (p.T).dot(q)[0, 0]
        alpha = v/m
        x += alpha*p
        r -= alpha*q
        v_pos = (r.T).dot(r)[0, 0]
        beta = v_pos/v
        p = r + beta*p
        v = v_pos
        k += 1
    return x
```

```python
def gauss_inv(A, b):
    N = A.shape[0]
    A_inv = eye(N)
    for i in range(N):
        div = A[i, i]
        for k in range(N):
            A[i, k] /= div
            A_inv[i, k] /= div
        for l in range(N):
            if l != i:
                mult = A[l, i]
                for m in range(N):
                    A[l, m] -= mult*A[i, m]
                    A_inv[l, m] -= mult*A_inv[i, m]
    x = A_inv.dot(b)
    return x
```

```python
def backsubs_gauss(A, b):
    N = b.size
    x = empty(shape=(N, 1), dtype=float)
    for m in range(N-1, -1, -1):
        x[m] = b[m]
        for i in range(m+1, N):
            x[m] -= A[m, i]*x[i]
    return x

def gauss_elim(A, b):
    N = b.size
    for m in range(N):
        div = A[m, m]
        for l in range(N):
            A[m, l] /= div
        b[m] /= div
        for i in range(m+1, N):
            mult = A[i, m]
            for k in range(N):
                A[i, k] -= mult*A[m, k]
            b[i] -= mult*b[m]
    x = backsubs_gauss(A, b)
    return x

def gauss_elim_changerow(A, b):
    N = b.size
    for m in range(N):
        # Pivoting:
        if A[m, m] == 0:
            div = 0
            for i in range(m+1, N):
                if (abs(A[i, i]) > abs(div)):
                    div = A[i, i]
                    ind = i
            # Changing rows:
            for k in range(N):
                A[m, k], A[i, k] = A[i, k], A[m, k]
            b[m], b[i] = b[i], b[m]
        else:
            div = A[m, m]
        # Gaussian elimination:
        A[m, :] /= div
        b[m] /= div
        for i in range(m+1, N):
            mult = A[i, m]
            A[i, :] -= mult*A[m, :]
            b[i] -= mult*b[m]
    x = backsubs_gauss(A, b)
    return x
```

```python
def tri_diagonal(A, b):
    N = b.size
    for i in range(N-1):
        A[i, i+1] /= A[i, i]
        b[i] /= A[i, i]
        A[i+1, i+1] -= A[i+1, i]*A[i, i+1]
        b[i+1] -= A[i+1, i]*b[i]
    b[N-1] /= A[N-1, N-1]
    x = empty(shape=(N, 1), dtype=float)
    x[N-1] = b[N-1]
    for i in range(N-2, -1, -1):
        x[i] = b[i] - A[i, i+1]*x[i+1]
    return x

def TDMA(A, b):
    n = b.size
    c_temp = zeros(shape=(n, 1), dtype=float)
    d_temp = zeros(shape=(n, 1), dtype=float)
    x = zeros(shape=(n, 1), dtype=float)
    c_temp[0, 0] = A[0, 1] / A[0, 0]
    d_temp[0, 0] = b[0, 0] / A[0, 0]
    for i in range(1, n-1):
        c_temp[i, 0] = A[i, i+1] / (A[i, i] - A[i, i-1] * c_temp[i-1, 0])
        d_temp[i, 0] = (b[i, 0] - A[i, i-1] * d_temp[i-1, 0]) / \
            (A[i, i] - A[i, i-1] * c_temp[i-1, 0])
    d_temp[n-1, 0] = (b[n-1, 0] - A[n-1, n-2] * d_temp[n-2, 0]) / \
        (A[n-1, n-1] - A[n-1, n-2] * c_temp[n-2, 0])
    x[n-1, 0] = d_temp[n-1, 0]
    for i in range(n-2, -1, -1):
        x[i, 0] = d_temp[i, 0] - c_temp[i, 0] * x[i+1, 0]
    return x
```

```python
def forw_subs_LU(L, b):
    N = b.size
    y = empty(shape=(N, 1), dtype=float)
    # Solving Ly = b:
    for k in range(N):  # k = 0, 1, ..., N-1
        s = 0
        for l in range(k):  # l = 0, 1, ..., N-2
            s += L[k, l]*y[l]
        y[k] = (b[k] - s) / L[k, k]
    return y

def back_subs_LU(U, y):
    N = y.size
    x = empty(shape=(N, 1), dtype=float)
    # Solving Ux = y:
    for k in range(N-1, -1, -1):  # k = N-1, N-2, ..., 0
        s = 0
        for l in range(k+1, N):  # l = k+1, k+2, ..., N-2, N-1
            s += U[k, l]*x[l]
        x[k] = (y[k] - s) / U[k, k]
    return x

def LU_decom(A, b):
    N = b.size
    U = array(A, copy=True)  # U = A
    L = eye(N, dtype=float)  # Identity I_N
    # Building L, U:
    for j in range(N-1):  # j = 0, 1, ..., N-2
        if U[j, j] == 0:
            print('There is no LU decomposition for this matrix')
            break
        else:
            for i in range(j+1, N):
                mult = U[i, j]/U[j, j]
                for k in range(N):  # k = 0, 1, ..., N-1
                    U[i, k] -= mult*U[j, k]
                    L[i, k] += mult*L[j, k]
    # Solving Ly=b:
    y = forw_subs_LU(L, b)
    # Solving Ux = y
    x = back_subs_LU(U, y)
    return x

def LU_decom_pvt(A, b):
    N = b.size
    # Partial pivoting:
    P = eye(N, dtype=float)  # Pivoting matrix
    for l in range(N):  # l = 0, 1, ..., N-1
        for i in range(l+1, N):  # i = l+1, l+2, ..., N-1
            major = A[l, 0]
            ind = l
            if abs(A[i, 0]) > abs(major):
                major = A[i, 0]
                ind = i
        aux = array(A[l, :], copy=True)
        pux = array(P[l, :], copy=True)
        A[l, :], A[ind, :] = array(A[ind, :], copy=True), aux
        P[l, :], P[ind, :] = array(P[ind, :], copy=True), pux
        issorted = False
        column_sorted = array(sorted(array(A[:, 0], copy=True), reverse=True))
        for m in range(N):
            if A[m, 0] == column_sorted[m]:
                issorted = True
        if issorted == True:
            break
    b = P.dot(b)  # Pb
    U = array(A, copy=True)  # U = A
    L = eye(N, dtype=float)  # Identity I_N
    # Building L, U:
    for j in range(N-1):  # j = 0, 1, ..., N-2
        if U[j, j] == 0:
            print('There is no LU decomposition for this matrix')
            break
        else:
            for i in range(j+1, N):
                mult = U[i, j]/U[j, j]
                for k in range(N):  # k = 0, 1, ..., N-1
                    U[i, k] -= mult*U[j, k]
                    L[i, k] += mult*L[j, k]
    # Solving Ly=b:
    y = forw_subs_LU(L, b)
    # Solving Ux = y
    x = back_subs_LU(U, y)
    return x
```

```python
def gauss_jacobi(A, b, x, target):
    N = b.size
    N_max = 10000
    residue = 1.0
    count = 0
    while (residue > target) and (count < N_max):
        for i in range(N):
            s = 0
            for j in range(N):
                if j != i:
                    s += A[i, j]*x[j, 0]
            x[i, 0] = (b[i, 0] - s) / A[i, i]
        residue = norm(b - A.dot(x))
        count += 1
    return x, count

def gauss_seidel(A, b, x, target):
    N = b.size
    N_max = 10000
    residue = 1.0
    count = 0
    while (residue > target) and (count < N_max):
        for i in range(N):
            s1 = 0
            x_new = array(x, copy=True)
            for j1 in range(i):  # New approximations
                s1 += A[i, j1]*x_new[j1, 0]
            s2 = 0
            for j2 in range(i+1, N):  # Old approximations
                s2 += A[i, j2]*x[j2, 0]
            x[i, 0] = (b[i, 0] - s1 - s2) / A[i, i]
        residue = norm(b - A.dot(x))
        count += 1
    return x, count
```

```python
def steepest_descent(A, b, x, target):
    r = b - A.dot(x)
    p = array(r, copy=True)
    N_max = 10000
    count = 0
    while (norm(r) > target) and (count < N_max):
        q = array(A.dot(p), copy=True)
        alpha = float((p.T).dot(r)/((p.T).dot(q)))
        x += alpha*p
        r -= alpha*q
        p = array(r, copy=True)
        count += 1
    return x, count

def conjugate_gradient(A, b, x, target):
    r = b - A.dot(x)
    p = array(r, copy=True)
    v = (r.T).dot(r)[0, 0]
    N_max = 10000
    count = 0
    while (norm(r) > target) and (count < N_max):
        q = A.dot(p)
        m = (p.T).dot(q)[0, 0]
        alpha = v/m
        x += alpha*p
        r -= alpha*q
        v_pos = (r.T).dot(r)[0, 0]
        beta = v_pos/v
        p = r + beta*p
        v = v_pos
        count += 1
    return x, count
```
