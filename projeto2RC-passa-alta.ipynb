```python
import numpy as np
import matplotlib.pyplot as plt
```

# Runge-Kutta de 4Âª ordem

```python
def RK4(a,b,h,r0,f):
  tpoints = np.arange(a,b,h)
  rpoints = []

  for t in tpoints:
    rpoints.append(r0)
    k1 = h*f(r0,t)
    k2 = h*f(r0 + 0.5*k1,t + 0.5*h)
    k3 = h*f(r0 + 0.5*k2,t + 0.5*h)
    k4 = h*f(r0+k3,t+h)
    r0 += (k1+2*k2+2*k3+k4)/6
  return tpoints, rpoints
```

# RC - passa alta

```python
def Vin(t):
  return (Vpp/2)*np.sin(2*np.pi*freq*t)

def dVin(t):
  return (Vpp/2)*2*np.pi*freq*np.cos(2*np.pi*freq*t)

def RCcircuit(Vout,t):
  return dVin(t) - Vout/(R*C)
```

# RC - Sinal temporal

```python
R = 500  #ohms
C = 1e-6 #faraday
Vpp = 3
freq = 1/(2*np.pi*R*C)
a = 0    #s
b = 3/freq  #s
N = 1000
h = (b - a)/N

tpoints = np.arange(a,b,h)
Vinpoints = Vin(tpoints)

rescaledt = [1000*t for t in tpoints]

plt.plot(rescaledt, Vinpoints, label='input signal')

#####################################################################################################################################################################

tpoints, Voutpoints = RK4(a,b,h,0,RCcircuit)

plt.plot(rescaledt, Voutpoints, label='output signal')
plt.title('Time response - RC high-pass - cutoff frequency')
plt.xlabel('time (ms)')
plt.ylabel('voltage (V)')
plt.legend(loc='upper right')
plt.minorticks_on()
plt.grid()
plt.savefig('RC-passa-alta-resp-tcof.png', format='png')
plt.show()
```

# RC - Sinal em frequÃªcia

```python
# constantes

Vpp = 3
a = 0    #s
R = 200  #ohms
C = 1e-6 #faraday
```

```python
freqpoints = np.logspace(1,4,50)
Vfpoints = []

for freq in freqpoints:

  b = 2/freq
  N = 1000
  h = (b - a)/N

  tpoints = np.arange(a,b,h)
  _, Voutpoints = RK4(a,b,h,0,RCcircuit)

  Vfpoints.append(max(Voutpoints) - min(Voutpoints))
```

```python
plt.title('Peak to peak output voltage in frequency domain - RC high-pass')
plt.xlabel('frequency (Hz)')
plt.ylabel(r'$V_{pp}$ (V)')
plt.grid(which='both')
plt.semilogx(freqpoints,Vfpoints)
plt.savefig('RC-passa-alta-resp-frequencia.png', format='png')
plt.show()
```

# RC - CapacitÃ¢ncia variante

```python
# constantes

Vpp = 3
a = 0    #s
R = 200  #ohms

colors = plt.cm.gnuplot(np.linspace(0, 1, 7))

for C, i in zip([1e-7, 5e-7, 1e-6, 5e-6, 1e-5, 5e-5, 1e-4], range(7)):
  freqpoints = np.logspace(1,7,100)
  Vfpoints = []

  for freq in freqpoints:
    b = 2/freq
    if C > 1.1e-7:
      N = 1000
    else:
      N = 10000
    h = (b - a)/N
    tpoints, Voutpoints = RK4(a,b,h,0,RCcircuit)
    Vfpoints.append(max(Voutpoints) - min(Voutpoints))

  plt.semilogx(freqpoints,Vfpoints,label=f'capacitance = {C:.0E}F', color=colors[i])

plt.title('Variant capacitance - RC high-pass')
plt.xlabel('frequency (Hz)')
plt.ylabel(r'$V_{pp}$ (V)')
plt.grid(which='both')
plt.legend()
plt.savefig('RC-passa-alta-cap-var.png', format='png')
plt.show()
```

# RC - ResistÃªncia variÃ¡vel

```python
# constantes

Vpp = 3
a = 0    #s
C = 1e-6 #F

colors = plt.cm.gnuplot(np.linspace(0, 1, 6))

for R, i in zip([10, 100, 500, 1000, 5000, 10000], range(6)):
  freqpoints = np.logspace(1,7,100)
  Vfpoints = []

  for freq in freqpoints:
    b = 2/freq
    if R < 100:
      N = 10000
    else:
      N = 1000
    h = (b - a)/N
    tpoints, Voutpoints = RK4(a,b,h,0,RCcircuit)
    Vfpoints.append(max(Voutpoints) - min(Voutpoints))

  plt.semilogx(freqpoints,Vfpoints,label=f'resistance = {R}$\Omega$', color=colors[i])

plt.title('Variant resistance - RC high-pass')
plt.xlabel('frequency (Hz)')
plt.ylabel(r'$V_{pp}$ (V)')
plt.grid(which='both')
plt.legend()
plt.savefig('RC-passa-alta-res-var.png', format='png')
plt.show()
```
